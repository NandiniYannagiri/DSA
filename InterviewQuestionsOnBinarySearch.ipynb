```python
## 1.Return The first index of Infinite Value
## Time complexity (O(n) - (adding all the elements from arr to arr1)) + (O(log n) - (for binary search)) = O(n)
## Function Call
def BinarySearchElement(arr1,i,j,x):
    if i<=j:
        mid = i+(j-i)//2
        if(arr1[mid] == x):
            return mid
        elif arr1[mid]<x:
            return BinarySearchElement(arr1,mid+1,j,x) # or use i=mid+1 , keep it in mind, use while loop there
        else:
            return BinarySearchElement(arr1,i,mid-1,x) # or use j=mid-1 ,  keep it in mind, use while loop there
    return -1

## Driver Code
arr = [1,32,5,3,-6,34,-50,10,1,-8,12,'inf','inf','inf','inf']
arr1 = []
for i in arr:
    arr1.append(float(i))
x = float('inf')
i = 0
j = len(arr)-1
result = BinarySearchElement(arr1,i,j,x)
print(result)
```
