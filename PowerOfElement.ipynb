```python
## Function Definition
def findPowerofElement(a,n):
    if n==0:
        return 1
    elif n<0:
        if n==-1:
            return 1/a
        else:
            n=abs(n)
            mid = n//2
            b = findPowerofElement(a,mid)
            result = (1/b)*(1/b)
        if abs(n)%2==0:
            return result
        return result*1/a
    elif n == 1:
        return a
    else:
        mid = n//2
        b = findPowerofElement(a,mid)
        result = b*b
    if n%2 == 0:
        return result
    else:
        return result*a
## Driver Code
a = 2
n = -2
## Function Call
result = findPowerofElement(a,n)
print(result)
```
