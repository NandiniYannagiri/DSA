```python
## Function Definition
## Time Complexity O(n)
## Space Complexity O(1) - because no extra space is used
def SearchElement(arr,x):
    for i in range(len(arr)):
        if x==arr[i]:
            return arr[i]
    return -1
## Driver code
arr = [20,11,2,54,7,40,1]
x = 2
## Function Call
result = SearchElement(arr,x)
print(result)
```
