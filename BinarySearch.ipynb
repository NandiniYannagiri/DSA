```python
## Time Complexity O(log n)
## Space complecity O(1) -- because no extra space is using
## Function Definition
def BinarySearch(arr,i,j,x):
    while i<=j:
        mid = i + (j-i)//2
        if arr[mid] == x:
            return x
        elif arr[mid] < x:
            return BinarySearch(arr,mid+1,j,x) # Or use normal way instead of recurtion ---> i = mid+1
        else:
            return BinarySearch(arr,i,mid-1,x) # Or use normal way instead of recurtion  ---> j = mid
    return -1

## Driver Code
arr = [2,14,33,76,1,34,65,9,21,7]
x = 9
i = 0
j = len(arr)-1
## Function Call
result = BinarySearch(arr,i,j,x)
print(result)
```
